```python
import time

def get_sentiment_score_A(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=50,
        n=1,
        stop=None,
        temperature=0.7,
    )
    sentiment_text = response.choices[0].text.strip()

    # Extract the last word, which should be the sentiment score
    last_word = sentiment_text.split()[-1]

    # Check if the last word is a valid float, otherwise return 0
    try:
        sentiment_score = float(last_word)
    except ValueError:
        sentiment_score = 0

    return sentiment_score


input_sentences = data[:125].tolist()

sentiment_scores = []

batch_size = 60
num_batches = (len(input_sentences) + batch_size - 1) // batch_size

Start_time = time.time()

for batch in range(num_batches):
    start = batch * batch_size
    end = min(start + batch_size, len(input_sentences))
    
    for sentence in input_sentences[start:end]:
        prompt = f"Please provide a sentiment score between -1 (negative) and 1 (positive) for the following sentence (just output a number for each sentence):\n\n{sentence}\n\n"
        score = get_sentiment_score_A(prompt)
        sentiment_scores.append(score)
    
    # Sleep between batches to avoid hitting the rate limit
    if batch < num_batches - 1:
        time.sleep(61)  # Sleep for 61 seconds to account for the time taken by the API calls

        
End_time = time.time()
        
Runtime = End_time - Start_time

print(f"Runtime: {Runtime:.2f} seconds")        
print(sentiment_scores)
```
